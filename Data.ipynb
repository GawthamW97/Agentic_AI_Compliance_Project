```python
import pandas as pd
```

```python
def build_hierarchy(row):
    return f"Section {row['section']} - Chapter - {str(row["hscode"]).strip("'")[:2]} - {row['description']}"
```

```python
df = pd.read_csv("./dataset/harmonized-system.csv")
df["text"] = df.apply(build_hierarchy,axis=1)
df.value_counts("text")
```
